Questão 1 - Altere nossa implementação de kNN para que ela possa calcular regressão. Dica: crie uma função predict_regression baseada na função predict_classification do documento notebook 02-Aula kNN.ipynb.

```python
def euclidean_distances(X, X_row):
    X_ = (X - X_row) ** 2
    return np.sum(X_, axis=1) ** 0.5
```

```python
def get_neighbors(X_train, test_row, k):
    distances = euclidean_distances(X_train, test_row)
    idx_sort = np.argsort(distances)
    return idx_sort[:k]
```

```python
def predict_regression(X, y, test_row, k):
    idx_sort = get_neighbors(X, test_row, k)
    output_values = y[idx_sort] 
    regression = output_values.mean()
    # print('idx_sort:{}, output_values:{}, prediction:{}'.format(idx_sort, output_values, regression))
    return regression
```

Questão 2 - (a)

```python
rot1 = predict_regression(X_treino, y_treino, np.array([42, 142000]), 5)
print('Valor: ', rot1)
```
