1. Resolver a classificação do Iris dataset para k=3 e k=5 com essas funções que implementamos no seguinte documento colab. Usando 70% dos dados para treino e 30% para teste. Ou seja: não é para usar a implementação do scikit para o kNN, nem o split_train_test do scikit.

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

Lendo dataset

```python
df = pd.read_csv("Iris.csv")
```

Definindo conjunto de teste e de treinamento

```python
train = df.sample(frac=0.7, random_state=56)
test = df.drop(train.index)
```

Transformando para um array numpy

```python
classes = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
X_train = train.drop(columns=['Id', 'Species']).to_numpy()
y_train = train['Species'].apply(classes.index).to_numpy()
```

```python
X_test = test.drop(columns=['Id', 'Species']).to_numpy()
```

```python
def euclidean_distances(X, X_row):
    X_ = (X - X_row) ** 2
    return np.sum(X_, axis=1) ** 0.5
```

```python
def get_neighbors(X_train, test_row, k):
    distances = euclidean_distances(X_train, test_row)
    idx_sort = np.argsort(distances)
    return idx_sort[:k]
```

```python
def predict_classification(X, y, test_row, k):
    idx_sort = get_neighbors(X, test_row, k)
    output_values = y[idx_sort]
    counts = np.unique(output_values, return_counts=True)
    idx_max = np.argmax(counts[1])
    prediction = counts[0][idx_max]
    # print('idx_sort:{}, output_values:{}, prediction:{}'.format(idx_sort, output_values, prediction))
    return prediction 
```

```python
def knn(X_train, y_train, X_test, k):
    lista = []
    for v in X_test:
        lista.append(predict_classification(X_train, y_train, v, k))
    return lista
```

Classificando com k=3

```python
test['classification_3'] = knn(X_train, y_train, X_test, 3)
test['classification_3'] = test['classification_3'].map({0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'})
iguais3 = test[test['classification_3']==test['Species']]
acuracy = iguais3.shape[0]/test['Species'].shape[0]
print("A acurácia para k = 3 foi de :", acuracy*100, "%")
```

Classificando com k = 5

```python
test['classification_5'] = knn(X_train, y_train, X_test, 5)
test['classification_5'] = test['classification_5'].map({0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'})
iguais5 = test[test['classification_5']==test['Species']]
acuracy = iguais5.shape[0]/test['Species'].shape[0]
print("A acurácia para k = 5 foi de :", acuracy*100, "%")
```
