### Questão 2.3
2.3. Usando somente o conjunto de treino, mostre as médias dos valores de cada feature para cada classe.


```python
# Calculando as médias das features da Iris-versicolor
vers_sepa_length = train[train['Species']=="Iris-versicolor"]["SepalLengthCm"].mean()
vers_sepa_width = train[train['Species']=="Iris-versicolor"]["SepalWidthCm"].mean()
vers_peta_length = train[train['Species']=="Iris-versicolor"]["PetalLengthCm"].mean()
vers_peta_width = train[train['Species']=="Iris-versicolor"]["PetalWidthCm"].mean()

# Calculando as médias das features da Iris-setosa
setosa_sepa_length = train[train['Species']=="Iris-setosa"]["SepalLengthCm"].mean()
setosa_sepa_width = train[train['Species']=="Iris-setosa"]["SepalWidthCm"].mean()
setosa_peta_length = train[train['Species']=="Iris-setosa"]["PetalLengthCm"].mean()
setosa_peta_width = train[train['Species']=="Iris-setosa"]["PetalWidthCm"].mean()

# Calculando as médias das features da Iris-virginica
virginica_sepa_length = train[train['Species']=="Iris-virginica"]["SepalLengthCm"].mean()
virginica_sepa_width = train[train['Species']=="Iris-virginica"]["SepalWidthCm"].mean()
virginica_peta_length = train[train['Species']=="Iris-virginica"]["PetalLengthCm"].mean()
virginica_peta_width = train[train['Species']=="Iris-virginica"]["PetalWidthCm"].mean()
print('Iris-versicolor')
print('Sepal Length: ', vers_sepa_length)
print('Sepal Width: ', vers_sepa_width)
print('Peta Length: ', vers_peta_length)
print('Peta Width: ', vers_peta_width)

print('\n Iris-setosa')
print('Sepal Length: ', setosa_sepa_length)
print('Sepal Width: ', setosa_sepa_width)
print('Peta Length: ', setosa_peta_length)
print('Peta Width: ', setosa_peta_width)

print('\n Iris-virginica')
print('Sepal Length: ', virginica_sepa_length)
print('Sepal Width: ', virginica_sepa_width)
print('Peta Length: ', virginica_peta_length)
print('Peta Width: ', virginica_peta_width)
```

#### Questão 2.4
2.4. Crie um algoritmo seu (sem usar aprendizado de máquina) para baseado nas médias das features calculadas no item 2.3. ou mesmo usando outras ideias suas, classificar corretamente as amostras do conjunto de teste. Aperfeçoe o seu algoritmo de forma a obter melhores resultados. Quanto por cento no conjunto de teste você conseguiu acertar corretamente o rótulo (classe) no seu melhor algoritmo, ou seja, qual a acurácia?


1. A ideia é verificar quanto vale a média das features de cada tipo de flor;
2. Depois verificar o quanto as features de uma nova se "distancia" dessas médias. Através de uma soma de todas as distâncias;
3. A soma cuja distância é menor será a escolhida para ser classificada;



```python
# Criando um nova feature "is_vers", para indicar o quão distante as amostras do conjunto de teste está das médias da flor Versicolor
test['is_vers'] = (abs(test['SepalLengthCm'] - vers_sepa_length))+(abs(test['SepalWidthCm'] - vers_sepa_width))+(abs(test['PetalLengthCm'] - vers_peta_length))+(abs(test['PetalWidthCm'] - vers_peta_width))

# Criando um nova feature "is_setosa", para indicar o quão distante as amostras do conjunto de teste está das médias da flor Setosa
test['is_setosa'] = (abs(test['SepalLengthCm'] - setosa_sepa_length))+(abs(test['SepalWidthCm'] - setosa_sepa_width))+(abs(test['PetalLengthCm'] - setosa_peta_length))+(abs(test['PetalWidthCm'] - setosa_peta_width))

# Criando um nova feature "is_virginica", para indicar o quão distante as amostras do conjunto de teste está das médias da flor Virginica
test['is_virginica'] = (abs(test['SepalLengthCm'] - virginica_sepa_length))+(abs(test['SepalWidthCm'] - virginica_sepa_width))+(abs(test['PetalLengthCm'] - virginica_peta_length))+(abs(test['PetalWidthCm'] - virginica_peta_width))
```

```python
# Definindo condição para descobrir qual das features tem o menor valor (is_vers | is_setosa | is_virginica)
condition1 = ((test['is_vers'] <=  test['is_setosa']) & (test['is_vers'] <= test['is_virginica']))
condition2 = ((test['is_setosa'] <=  test['is_vers']) & (test['is_setosa'] <= test['is_virginica']))
condition3 = ((test['is_virginica'] <=  test['is_setosa']) & (test['is_virginica'] <= test['is_vers']))

# A feature (is_vers | is_setosa | is_virginica) com menor valor definirá como a flor será classificada. Pois ela estará mais parecida com a médias calculadas.
test['new_classification'] = np.where(condition1, 'Iris-versicolor',(np.where(condition2,'Iris-setosa', np.where(condition3,'Iris-virginica',np.nan))))
```

```python
iguais = test[test['Species'] == test['new_classification']]
# Calculando a precisão
# A melhor precisão foi de 1
acuracy = iguais.shape[0]/test.shape[0]
print('A acurácia foi de ',acuracy*100,'%')
```
