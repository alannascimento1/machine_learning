```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

 O conjunto de dados de qualidade de vinho consiste na predição da qualidade de vinhos brancos sobre uma escala dada medições químicas de cada vinho. Baixe o conjunto de qualidade de vinho e faça o seguinte com ele em um Jupyter Notebook:

```python
df = pd.read_csv('/content/drive/MyDrive/AprendizadoDeMaquina/datasets/winequality-white.csv', sep=';')
df.shape
```

```python
X = df.drop(columns=['quality']).to_numpy()
y = df['quality'].to_numpy()
```

Separe 70% dos dados para treino (conjunto de treino) e 30% dos dados para teste (conjunto de teste).



```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

Crie um modelo de regressão de qualidade de vinho usando o kNN para k=5.

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_treino, y_treino)
```

Use o modelo de regressão sobre X_test para calcular os valores preditos dos rótulos (y_pred)

```python
y_pred = knn.predict(X_teste)
y_pred
```
