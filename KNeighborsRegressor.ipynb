```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv('/content/drive/MyDrive/AprendizadoDeMaquina/datasets/winequality-white.csv', sep=';')
df.shape
```

```python
X = df.drop(columns=['quality']).to_numpy()
y = df['quality'].to_numpy()
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_treino, y_treino)
```

```python
y_teste_new = knn.predict(X_teste)
```
