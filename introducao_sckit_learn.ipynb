```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

Questão 2

```python
df = pd.read_csv("winequality-white.csv", sep=';')
```

Questão 2.1 - Exiba a quantidade de amostras existente em cada classe de vinho, onde as classes são representadas por valores inteiros entre 0 e 10.

```python
print('Classe 3: ', df[df['quality']==3].shape[0])
print('Classe 4: ', df[df['quality']==4].shape[0])
print('Classe 5: ', df[df['quality']==5].shape[0])
print('Classe 6: ', df[df['quality']==6].shape[0])
print('Classe 7: ', df[df['quality']==7].shape[0])
print('Classe 8: ', df[df['quality']==8].shape[0])
print('Classe 9: ', df[df['quality']==9].shape[0])
```

```python
X = df.drop(columns=['quality']).to_numpy()
y = df['quality'].to_numpy()
```

2.2. Separe 70% dos dados para treino (conjunto de treino) e 30% dos dados para teste (conjunto de teste).

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

2.3. Crie diferentes modelos de classificação de vinho usando o kNN com diferentes valores de k.



```python
neighbors = np.arange(1, 30)
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)    
    knn.fit(X_treino, y_treino)
    test_accuracy[i] = knn.score(X_teste, y_teste)
```

2.4. Mostre os resultados de acurácia para os modelos criados. Qual o modelo com melhor acurácia?



```python
print('Vetor de acurácias: \n', test_accuracy)
print('\nO melhor modelo é com k: ' ,np.argmax(test_accuracy)+1)
print('Maior acurácia: ', test_accuracy[np.argmax(test_accuracy)])
```

```python
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
